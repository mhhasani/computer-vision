**difference between OpenCV and matplotlib in display image**

OpenCV is a library for computer vision and image processing, and it has a function called "imshow" that can display an image in a window. This function is useful for displaying images that are being processed in real-time, such as those from a camera or video stream. In addition to displaying the image, OpenCV also provides tools for image processing and analysis, such as filtering, edge detection, and object detection.

Matplotlib, on the other hand, is a data visualization library that can also display images. Matplotlib's "imshow" function is typically used for displaying static images, such as those that are loaded from a file or generated as output from a data analysis pipeline. Matplotlib provides a wide range of options for customizing the appearance of the image, such as adjusting the color map, adding annotations, and creating subplots.

In summary, while both OpenCV and Matplotlib can display images, OpenCV is more suited for real-time processing and analysis of images, while Matplotlib is more suited for static visualization and customization of images.

The ``shape`` attribute of an image in Python gives us information about the dimensions of the image. The output of ``image.shape`` is a tuple containing three elements: (height, width, channels).

- The first element, height, gives the height of the image in pixels.
- The second element, width, gives the width of the image in pixels.
- The third element, channels, gives the number of channels in the image. For grayscale images, this value is 1, and for color images, this value is usually 3 (for the Red, Green, and Blue color channels).
For example, if we have an RGB color image with dimensions 680 pixels by 1020 pixels, the output of image.shape would be (680, 1020, 3).

--**بخش ج**--

```python
os.makedirs('Q1/new_directory', exist_ok=True)
for image_file in os.listdir('Q1'):
    if image_file.endswith('.jpeg'):
        image = cv2.imread(os.path.join('Q1', image_file))
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image_name = os.path.splitext(image_file)[0]
        with open(os.path.join('Q1/new_directory', f'{image_name}.txt'), 'w') as f:
            height, width = gray_image.shape[:2]
            f.write(f'Height: {height}\n')
            f.write(f'Width: {width}\n')
        cv2.imwrite(os.path.join('Q1/new_directory', f'{image_name}.jpg'), gray_image)
```
